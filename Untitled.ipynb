```python
def owner_number(text):
    a=3
    if(text=='First'):
        a=1
        return a
    elif(text=='Second'):
        a=2
        return a
    else:
        return a
    
```

```python
import re

def extract_numbers(text):
    # Use regex to find all digits in the string and join them
    numbers = re.findall(r'\d+', text)
    # Join the list of numbers into a single string and convert to an integer
    return int(''.join(numbers)) if numbers else None
```

```python
car_data['Engine']=car_data['Engine'].apply(extract_numbers)
```

```python
car_data.head()
```

```python
 def extract_rpm(text):
    l = []
    count = 5  # This is the character count to skip, which may not be necessary
    i = 1

    for x in text:
        if i <= count:
            i += 1
        else:
            if x.isalpha():
                continue
            elif x==' ':
                continue
            elif x=='@':
                continue
            else:
                l.append(x)

    return int(''.join(l)) if l else None  
```

```python
car_data['Max Power']=car_data['Max Power'].apply(extract_rpm)
```

```python
#so this is very big error to predict the car data we can use other method also
from sklearn.ensemble import RandomForestRegressor
```

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = RandomForestRegressor(max_leaf_nodes=max_leaf_nodes, random_state=1)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
```

```python
def best_node(l):
    for max_leaf_nodes in l:
        count=4294967295
        i=0
        my_mae = get_mae(max_leaf_nodes, X_train, X_val, y_train, y_val)
        if(my_mae<count):
            count=my_mae
            i=max_leaf_nodes
    return i
    
```

```python
l=[5 , 10 , 20 , 50 , 100, 200 ,400 , 500 ,1000]
best_node=best_node(l)
```

```python
fcar_model=RandomForestRegressor(max_leaf_nodes=best_node, random_state=1)
```

```python
fcar_model.fit(X_train,y_train)
prediction=fcar_model.predict(X_val)
```

```python
fmae=mean_absolute_error(y_val , prediction)
print(fmae)
```
